```python
# add column "day_night"
df['day_night'] = df['time'].apply(lambda x: "night" if '00:00' <= x <= '06:00' or '18:00' <= x <= '24:00' else 'day')
df.head()
```

```python
# add column "weekday_weekend"
df["date"] = pd.to_datetime(df["date"])
df['weekday'] =  df['date'].dt.weekday
df["IsWeekend"] = df['date'].dt.weekday > 5
```

```python
# add column "month"
df['month'] = df['date'].dt.month
```
